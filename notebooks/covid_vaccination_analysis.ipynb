# Create Jupyter Notebook

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for visualizations
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_theme()
```

## Data Collection

```python
# Download COVID-19 vaccination data from Our World in Data
url="https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations.csv"
df=pd.read_csv(url)

# Display basic information about the dataset
print("Dataset Info:")
print(df.info())
print("\nFirst few rows:")
print(df.head())
```

## Data Cleaning

```python
# Handle missing values
df['total_vaccinations'] = df['total_vaccinations'].fillna(0)
df['people_vaccinated'] = df['people_vaccinated'].fillna(0)
df['people_fully_vaccinated'] = df['people_fully_vaccinated'].fillna(0)

# Convert date column to datetime
df['date'] = pd.to_datetime(df['date'])

# Create a clean subset of data for top 10 countries by total vaccinations
top_countries = df.groupby('location')['total_vaccinations'].max().sort_values(ascending=False).head(10).index
df_top10 = df[df['location'].isin(top_countries)]
```

## Exploratory Data Analysis (EDA)

```python
# Create a visualization of vaccination progress for top 10 countries
plt.figure(figsize=(15, 8))
for country in top_countries:
    country_data = df_top10[df_top10['location'] == country]
    plt.plot(country_data['date'], country_data['total_vaccinations'], label=country)

plt.title('COVID-19 Vaccination Progress in Top 10 Countries')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# Advanced Analysis

```python
# Calculate vaccination rates
df_top10 = df[df['location'].isin(top_countries)].copy()
df_top10['vaccination_rate'] = df_top10['people_fully_vaccinated'] / df_top10['total_vaccinations']

# Create a box plot of vaccination rates by country
plt.figure(figsize=(12, 6))
sns.boxplot(x='location', y='vaccination_rate', data=df_top10)
plt.xticks(rotation=45)
plt.title('Distribution of Vaccination Rates by Country')
plt.tight_layout()
plt.show()

# Calculate daily vaccination rates
df_top10['daily_vaccinations_per_million'] = df_top10.groupby('location')['total_vaccinations'].diff()
```
